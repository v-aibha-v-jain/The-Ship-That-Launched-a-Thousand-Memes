```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("train.csv")
```

```python
# some basic information about the dataset
print("Basic Info for the dataset:")
df.info()
```

```python
# To check the number of null cells in the dataset
print("Null cell stats of the dataset:")
df.isnull().sum()
```

```python
# To check the number of duplicate data in the dataset
print("duplicate cell stats of the dataset:")
df.duplicated().sum()
```

```python
# As there are no duplicates in this dataset we don't need to use drop_duplicates() to delete duplicate rows
```

```python
# Sample data of this csv file
print("Sample data of the dataset:")
df.sample(5)
```

```python
# Summary statistics
print("\nSummary Statistics for the 'train' dataset:")
df.describe()
```

```python
# Total number of people survived
print("Number of people survived")
df[["Survived"]].sum()
```

```python
# Number of females and males survived
df.groupby(['Sex', 'Survived'])['Survived'].count()
```

```python
# Visualizations
# 1. Survived Countplot
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived')
plt.title('Survived Count')
plt.show()
```

```python
# 2. Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# 3. Fare Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Fare', bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# 4. Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex')
plt.title('Gender Distribution')
plt.show()
```

```python
# 5. Passenger Class (Pclass) Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# 6. Siblings/Spouses (SibSp) Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='SibSp')
plt.title('Siblings/Spouses Distribution')
plt.show()
```

```python
# 7. Parents/Children (Parch) Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Parch')
plt.title('Parents/Children Distribution')
plt.show()
```

```python
# 8. Port of Embarkation (Embarked) Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Embarked')
plt.title('Port of Embarkation Distribution')
plt.show()
```

```python
# Summary of Findings
# - The 'Survived' countplot shows the distribution of passengers who survived (0 = Not Survived, 1 = Survived).
# - The 'Age' distribution indicates the age distribution of passengers.
# - The 'Fare' distribution shows the fare distribution.
# - The 'Gender' distribution displays the count of males and females.
# - The 'Pclass' distribution displays the count of total number of people in 1st, 2nd and 3rd class
# - The 'sibsp' distribution distalys the number of people having total number of Siblings/Spouses
# - The 'parch' distribution displays the number of parent/children by each person
# - The 'Embarked' distribution dislplays the number of people at different port
```
